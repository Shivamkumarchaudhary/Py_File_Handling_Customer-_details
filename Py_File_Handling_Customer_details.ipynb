### Store customer details in a file, then perform following operations.
1.Add new record

2.Search record

3.Delate record

4.Update record

5.Display record

```python
while (True):
    print("Welcome to customer portal")
    print("1. Add new customer")
    print("2. Delate customer")
    print("3. Update customer")
    print("4. Search customer")
    print("5. Show all customer")
    print("6. Exit customer")
    ch = int(input("Enter your choice: "))
    if ch==1:
        def add():
            cid = input("Customer ID : ")
            nm = input("Customer Name : ")
            ad = input("Customer Address : ")
            cn = input("Contact Number : ")
    
            f = open("file_path.txt","a")
            f.write(cid+"\t")
            f.write(nm+"\t")
            f.write(ad+"\t")
            f.write(cn+"\n")
            print("Record Added")
            f.close()
    
        add()
        continue
        
    if ch==2:
        def delate():
            i = input("Enter the ID to remove from file")
            with open("file_path.txt","r") as f:
                all = f.readlines()
            
            with open("file_path.txt","w") as f:
                for data in all:
                    d = data.split("\t",1)
                    if(d[0] != i):
                        f.writelines(data)
            
                        print("Recorded Delated")
        delate()
        continue
    if ch==3:
        def update():
            i = input("Enter the ID to be Update from file")
            with open("file_path.txt","r") as f:
                all = f.readlines()
            with open ("file_path.txt","w") as f:
                for data in all:
                    d = data.split("\t",1)
                    if(d[0]==i):
                        nn = input("New Name")
                        na = input("New Address")
                        nc = input("New contact Number")
                        f.writelines(d[0] + "\t" + nn + "\t" + na +"\t" + nc)
                    else:
                        f.writelines(data)
                        print("Record Updated")
        update()
        continue
    if ch==4:
        def search():
            i = input("Enter the ID to be Searched from files")
            with open("file_path.txt","r") as f:
                all = f.readlines()
                for data in all:
                    d = data .split("\t",1)
                    if (d[0]==i):
                        print(data)
        search()
        continue
    if ch==5:
        def show():
            f = open("file_path.txt","r")
            print("ID \t Name \t Address \t Contact no")
            print(f.read())
            f.close()
        show()
        continue
    if ch==6:
        break
         
```
